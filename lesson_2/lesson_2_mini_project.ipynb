```python
bookings['total_kids'] = bookings.children + bookings.babies
```

```python
bookings['total_kids']
```

```python
total_kids = bookings.groupby('hotel', as_index=False)  \
                     .agg({'total_kids': 'mean'})  \
                     .round(2)
```

```python
total_kids
```

```python
bookings['has_kids'] = bookings.total_kids != 0
```

```python
has_kids = bookings.query('has_kids == True')
```

```python
has_kids_canceled = bookings.query('has_kids == True and is_canceled == 1')
```

```python
hasnt_kids = bookings.query('has_kids == False')
```

```python
hasnt_kids_canceled = bookings.query('has_kids == False and is_canceled == 1')
```

```python
churn_rate_hk = has_kids_canceled.shape[0] / has_kids.shape[0]
round(churn_rate_hk, (4))
```

```python
churn_rate_dhk = hasnt_kids_canceled.shape[0] / hasnt_kids.shape[0]
round(churn_rate_dhk, (4))
```

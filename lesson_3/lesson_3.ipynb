```python
def read_n_agg(ptf):
    df = pd.read_csv(ptf,sep=';')
    df = df.groupby('company').agg({'income': 'mean'})
    return df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
taxi = pd.read_csv('3_taxi_peru.csv', parse_dates=['start_at', 'end_at', 'arrived_at'], sep=';')
```

```python
taxi
```

```python
taxi['start_at'] = pd.to_datetime(taxi['start_at'])
taxi['end_at'] = pd.to_datetime(taxi['end_at'])
taxi['arrived_at'] = pd.to_datetime(taxi['arrived_at'])
```

```python
taxi.groupby('start_type', as_index = False).agg({'start_at': 'count'})
```

```python
taxi.query('start_type != "delayed"').shape
```

```python
taxi['wait_time'] = taxi.arrived_at - taxi.start_at
```

```python
taxi['wait_time'] = taxi.wait_time.astype('timedelta64[m]')
```

```python
taxi_reserved = taxi.query('start_type == "reserved"')
```

```python
driver_delay = taxi_reserved.query('wait_time > 0.0')
```

```python
driver_delay.groupby('driver_id', as_index=False).agg({'wait_time': 'count'}).sort_values('wait_time', ascending=False)
```

```python
taxi['month'] = taxi.start_at.dt.month
```

```python
taxi['weekday'] = taxi.start_at.dt.strftime('%A')
```

```python
taxi
```

```python
sns.countplot(taxi['month'])
```

```python
sns.countplot(taxi['weekday'], order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
```

```python
unique_user = taxi.groupby('month', as_index = False).agg({'user_id': 'nunique'}).sort_values('month')
```

```python
unique_user
```

```python
sns.lineplot(unique_user.month, unique_user.user_id)
```

```python
source = taxi.groupby('source')  \
    .agg({'source': 'count'})  \
    .rename(columns={'source': 'count'})  \
    .reset_index()  \
    .sort_values('count', ascending=False)
```

```python
source
```

```python
sns.barplot(x='source', y='count', data=source)
```

```python
sns.countplot(taxi['icon'])
```

```python
sns.countplot(data=taxi, x='end_state', hue="source")
```

```python
driver_score_counts = taxi.driver_score  \
    .value_counts('normalize')  \
    .mul(100)  \
    .round(2)  \
    .reset_index()  \
    .rename(columns={'index': 'driver_score', 'driver_score': 'percentage'})  \
    .sort_values('driver_score')
```

```python
driver_score_counts
```

```python
ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage')
sns.despine()  # убрать часть рамки графика
plt.show()
```

```python
ride_score_counts = taxi.rider_score  \
    .value_counts('normalize')  \
    .mul(100)  \
    .round(2)  \
    .reset_index()  \
    .rename(columns={'index': 'rider_score', 'rider_score': 'percentage'})  \
    .sort_values('rider_score')
```

```python
ride_score_counts
```

```python
ax = sns.barplot(x='rider_score', y='percentage', data=ride_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Rider Score', ylabel='Percentage')
sns.despine()  # убрать часть рамки графика
plt.show()
```

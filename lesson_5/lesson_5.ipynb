```python
retail = retail.query('InvoiceNoCancel == False')
```

```python
df = retail.query('Country == "Germany"').groupby('CustomerID', as_index=False).agg({'InvoiceNo': 'nunique'})
```

```python
ger_quantile = df.InvoiceNo.quantile(q=0.8)
```

```python
df1 = df.query('InvoiceNo > @ger_quantile')
```

```python
germany_top = df1.CustomerID
```

```python
germany_top
```

```python
top_retail_germany = retail.query('CustomerID in @germany_top')
```

```python
top_retail_germany
```

```python
top_retail_germany.groupby('StockCode', as_index=False).agg({'Description': 'count'}).sort_values('Description')
```

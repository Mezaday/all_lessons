```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data_path = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-v-ostrouhov/shared/homeworks/python_ds_miniprojects/5_subsid/'
```

```python
df = pd.DataFrame()
```

```python
for root, _, files in os.walk(data_path):
    for file in files:
        if(file.startswith('tm')):
            filepath = os.path.join(root,file)
            df1 = pd.read_csv(filepath, sep=';')
            df = df.append(df1, ignore_index = True)
```

```python
df = df.dropna()
```

```python
df
```

```python
def rename(i):
    if i.startswith('id'):
        n = i
        return n
    else:
        n = 'id' + i 
        return n
```

```python
df['SUBS_ID'] = df['SUBS_ID'].apply(rename)
```

```python
df
```

```python
prod_activations_logs = pd.read_csv('/mnt/HC_Volume_18315164/home-jupyter/jupyter-v-ostrouhov/shared/homeworks/python_ds_miniprojects/5_subsid/subsid/prod_activations_logs.csv', sep=';')
```

```python
prod_activations_logs_merge = df.merge(prod_activations_logs, on=['SUBS_ID', 'PROD_ID'], how='inner')
```

```python
prod_activations_logs_merge
```

```python
prod_activations_logs_merge['ACT_DTTM'] = pd.to_datetime(prod_activations_logs_merge['ACT_DTTM'], format='%d-%m-%Y %H:%M')
```

```python
prod_activations_logs_merge['START_DTTM'] = pd.to_datetime(prod_activations_logs_merge['START_DTTM'], format='%d-%m-%Y %H:%M')
```

```python
prod_activations_logs_merge['END_DTTM'] = pd.to_datetime(prod_activations_logs_merge['END_DTTM'], format='%d-%m-%Y %H:%M')
```

```python
prod_activations_logs_merge['difference'] = prod_activations_logs_merge['END_DTTM'] - prod_activations_logs_merge['START_DTTM']
```

```python
prod_activations_logs_merge['Succes_DTTM'] = prod_activations_logs_merge['difference'] > '5m'
```

```python
prod_activations_logs_merge
```

```python
prod_activations_logs_merge = prod_activations_logs_merge.query('Succes_DTTM == True').sort_values('SUBS_ID')
```

```python
prod_activations_logs_merge.to_csv('df.csv', sep=';')
```

```python
ads_data_cv = ads_data.groupby(['ad_id', 'event'])  \
    .agg({'ad_id': 'count'})  \
    .rename(columns={'ad_id': 'count'})  \
    .reset_index()  \
    .pivot(index='ad_id', columns='event', values='count')  \
    .fillna(0)  \
    .reset_index()  \
```

```python
ads_data_cv_query = ads_data_cv.query('view > 0')
```

```python
ads_data_view = ads_data.query("event == 'view'")
```

```python
ads_date_view_mean = ads_data_view.groupby(['date', 'ad_id'], as_index=False)  \
    .agg({'event': 'count'})  \
    .groupby('date')  \
    .agg({'event': 'mean'})  \
    .reset_index()
```

```python
ads_date_view_mean
```

```python
ads_date_view_mean['roll'] = ads_date_view_mean.event.rolling(2).mean().fillna(0)
```

```python
ads_date_view_mean
```

```python
sns.lineplot(x=ads_date_view_mean.date, y = ads_date_view_mean.event)
sns.lineplot(x=ads_date_view_mean.date, y = ads_date_view_mean.roll)
```

```python
ads_data_full = ads_data.merge(ads_clients_data, on='client_union_id', how='right')
```

```python
ads_data_full
```

```python
ads_data_full.dtypes
```

```python
ads_data_full['date_x'] = pd.to_datetime(ads_data_full['date_x'])
```

```python
ads_data_full['time'] = pd.to_datetime(ads_data_full['time'])
```

```python
ads_data_full['date_y'] = pd.to_datetime(ads_data_full['date_y'])
```

```python
ads_data_full['create_date'] = pd.to_datetime(ads_data_full['create_date'])
```

```python
ads_data_full['date_time'] = ads_data_full.date_x - ads_data_full.create_date
```

```python
ads_data_full
```

```python
ads_data_full.groupby('client_union_id', as_index=False).agg({'date_time': 'min'}).mean()
```

```python
ads_data_full_ = ads_data_full.groupby('client_union_id', as_index=False).agg({'date_time': 'min'})
ads_data_full_
```

```python
pd.Timedelta(365, unit='day')
```

```python
ads_data_full_365d = ads_data_full_.query("date_time < '365d'")
```

```python
ads_data_full_365d
```

```python
ads_data_full_365d['first_ad'] = pd.cut(ads_data_full_365d['date_time'], bins=pd.to_timedelta(['0d', '30d', '90d', '180d', '365d']), labels = ['less than 30 day', '30-90 days', '90-180 days', 'more than 180 days'])
```

```python
ads_data_full_365d.first_ad.value_counts()
```
